Introduction to tuples : 

Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
# creating a tuple
empty_tuple = ()
print(empty_tuple)  # Output: ()
print(type(empty_tuple))  # Output: <class 'tuple'>
```

```python
lst = list()
print(type(lst))  # Output: <class 'list'>
tpl = tuple()
print(type(tpl))  # Output: <class 'tuple'>
```

```python
numbers = tuple([1, 2, 3, 4, 5])
numbers
```

```python
list((1, 2, 3, 4, 5))
```

```python
mixed_tuple = (1, "Hello", 3.14, [1, 2, 3], (4, 5))
mixed_tuple
```

```python
## Accessing Tuple Elements
numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
## tuples operations
concatenation_tuples = numbers + mixed_tuple
print(concatenation_tuples)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Packing and Unpacking
# Packing
packed_tuple = numbers
a, b, c, d, e = packed_tuple
print(a)  # Output: 1
print(b)  # Output: 2
print(c)  # Output: 3
print(d)  # Output: 4
print(e)  # Output: 5
```

```python
# unpacking
unpacked_tuple = numbers
a, b, c, d, e = unpacked_tuple
print(a, b, c, d, e)  # Output: 1 2 3 4 5
```
