1 . TO DO LIST

```python
to_do_list = ["Buy groceries", "Clean the house", "Pay bills", "Call mom"]
print(to_do_list[0])  # Output: Buy groceries
print(to_do_list[1])  # Output: Clean the house
```

```python
to_do_list.append("Go for a walk")
print(to_do_list)
```

```python
to_do_list.remove("Pay bills")
print(to_do_list)
```

```python
if "Buy groceries" in to_do_list:
    print("You need to buy groceries.")
```

```python
print("To do list remains unchanged:")
for task in to_do_list:
    print(task)
    
```
