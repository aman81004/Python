Dictionaries 
These are unordered collections of items. They store data in key-value pairs . Key must be unique and immutable(eg. strings ,numbers , tuples) , which values can be of any type

```python
## Creating Dictionaries
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
student = {"name": "Aman", "age": 20, "grade": "A"}
print(student)
print(type(student))
```

```python
## Accessing dictionary values
student = {"name": "Aman", "age": 20, "grade": "A"}
print(student["name"])
print(student["age"])
print(student["grade"])
```

```python
## Accessing using get() method
print(student.get('name'))
print(student.get("age"))
print(student.get("grade"))
```

```python
## Modifying dictionary elements
## Dictionaries are mutable, which means you can change their content after they have been created. You can add new key-value pairs, update existing values, or remove key-value pairs.
print(student)
```

```python
student["age"] = 21
print(student)
```

```python
student["address"] = "India"
print(student)
```

```python
del student["grade"]
print(student)
```

```python
## Dictionary Methods

keys = student.keys()
print(keys)
```

```python
values = student.values()
print(values)
```

```python
items = student.items()
print(items)
```

```python
## Shallow Copying of Dictionaries
student_copy = student.copy()
print(student_copy)
```

```python
student["name"] = "Tvisha"
print(student)
print(student_copy)
```
