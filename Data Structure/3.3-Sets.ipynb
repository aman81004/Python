Sets

These are built in data type in python used to store collections of unique items. They are unordered , meaning that the elements do not follow a specific order, and they do not allow duplicate elements . Sets are useful for memberships tests , eliminating duplicate entries , and performing mathematical set operations like union, intersection , difference , and symmetric difference .

```python
## create a set
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set([1, 2, 3, 4, 5])
print(my_empty_set)
print(type(my_empty_set))
```

```python
## Basic set operations
## Adding and Removing Elements
my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
## Removing elements 
my_set.remove(3)
print(my_set)
```

```python
my_set.discard(11)  # No error if element not found    
print(my_set)
```

```python
# pop method removes and returns an arbitrary element from the set. If the set is empty, it raises a KeyError.
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
## Set Membership test
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Set Operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# Union
union_set = A | B
print("Union:", union_set)

# Intersection
intersection_set = A & B
print("Intersection:", intersection_set)

# Difference
difference_set = A - B
print("Difference:", difference_set)

# Symmetric Difference
sym_diff_set = A ^ B    
print("Symmetric Difference:", sym_diff_set)

A.intersection_update(B)
print("A after intersection_update:", A)
```

```python
## Sets Methods
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# issubset
print("Is A a subset of B?", A.issubset(B))

# issuperset
print("Is A a superset of B?", A.issuperset(B))

```

```python
lt = [1, 2 , 2, 3, 4, 4, 5]
set(lt)
```

```python
## Counting unique elements in a text
text = "hello world"
unique_characters = set(text)
print(unique_characters)

```
