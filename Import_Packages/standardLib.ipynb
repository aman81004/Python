Standard Library

python's standard library is a vast collection of modules and packages that come bundled with Python , providing a wide range of functionalities out of the box . Here's an overview of some of the most commonly used modules nad packages in python standard library .

```python
import array
arr = array.array('i',[1,2,3,4])
print(arr)
```

```python
import math 
print(math.sqrt(16))
```

```python
import random 
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry']))
```

```python
### File and Directory Access

import os
print(os.getcwd()) ## Get current working directory
print(os.listdir()) ## List all files and directories in the current working directory
```

```python
os.mkdir('new_directory') ## Create a new directory
os.rename('new_directory','renamed_directory') ## Rename a directory
os.rmdir('renamed_directory') ## Remove a directory
```

```python
## High level file operations
import shutil
shutil.copy('source_file.txt','destination_file.txt') ## Copy a file
shutil.move('source_file.txt','destination_directory/') ## Move a file
```

```python
## Data serialization

import json
data = {'name':'John','age':30,'city':'New York'}
json_string = json.dumps(data) ## Convert Python object to JSON string
print(json_string)

parsed_data = json.loads(json_string) ## Convert JSON string back to Python object
print(parsed_data)
```

```python
## CSV
import csv
with open('data.csv',mode = 'w',newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Name','Age','City'])
    writer.writerow(['Alice',25,'Los Angeles'])
    writer.writerow(['Bob',30,'Chicago'])
    
with open('data.csv',mode='r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        print(row)
    
```

```python
## datetime
import datetime
now = datetime.datetime.now()
print(now)
print(now.year)
print(now.strftime('%Y-%m-%d %H:%M:%S'))

yesterday = now - datetime.timedelta(days=1)
print(yesterday)
```

```python
## Time

import time
print(time.time()) ## Current time in seconds since the epoch
print(time.ctime()) ## Current time as a human-readable string
time.sleep(2) ## Pause execution for 2 seconds
print("2 seconds have passed")
```

```python
## Rgular Expressions
import re
text = "The rain in Spain stays mainly in the plain."
matches = re.findall(r'\b\w+ain\b',text) ## Find all words ending with 'ain'
print(matches)
```
