Data Visualization with Seaborn 
Seaborn is Python visualization lib based on matplotlib that provides a high level interface for drawing attractive and informative statistical graphics . Seaborn helps in creating complex visualization with just few lines of codes .

```python
import seaborn as sns
```

```python
## Basic plotting with seaborn
tips = sns.load_dataset("tips")
tips
```

```python
## create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot( x = 'total_bill', y = 'tip', data = tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line plot

sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title("Line Plot of Size vs Total Bill")
plt.show()
```

```python
## Categorical plot
## bar plot

sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Bar Plot of Day vs Total Bill")
plt.show()
```

```python
## Box plot
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Box Plot of Day vs Total Bill")
plt.show()
```

```python
## violin plot

sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Violin Plot of Day vs Total Bill")
plt.show()
```

```python
## Histogram
sns.histplot(x = 'total_bill', data = tips, bins = 10, kde = True)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
## KDE plot
sns.kdeplot(x = 'total_bill', data = tips, fill = True)
plt.title("KDE Plot of Total Bill")
plt.show()
```

```python
## Pair plot

sns.pairplot(tips)
plt.show()
```

```python
## Heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
