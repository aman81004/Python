Pandas-DataFrame and Series
Pandas two primary data manipulation library in Python , widely used for data analysis and data cleaning . It provides two primary data structures : series and data frame . A series is 1-D array like object , while data frame is a 2-D , size mutable and potentially heterogeneous tabular data structure with labeled axes ( rows and coloumns )

```python
import pandas as pd
```

```python
## Series Creation

data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
```

```python
## Create a series from a dictionary
import pandas as pd
data_dict = {'a': 1, 'b': 2, 'c': 3}
series_from_dict = pd.Series(data_dict)
print(series_from_dict)
```

```python
data = [10, 20, 30, 40, 50]
index = ['A', 'B', 'C', 'D', 'E']

series_with_index = pd.Series(data, index=index)
print(series_with_index)
```

```python
## DataFrame Creation
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print(df)
```

```python
## Creating a DataFrame from a list of dictionaries

data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'}   
]

df_from_list = pd.DataFrame(data)
print(df_from_list)
```

```python
df = pd.read_csv('sales.csv')
print(df.head())
print(df.tail())
```

```python
## Accesing data from DataFrame
print(df['First Name']) ## Accessing a column
```

```python
df.loc[0] ## Accessing the first row using loc
```

```python
df.iloc[0] ## Accessing the first row using iloc
```

```python
## Accessiong a specific element
print(df.loc[0, 'First Name']) ## Using loc to access a specific element
print(df.iloc[0, 0]) ## Using iloc to access a specific element
```
