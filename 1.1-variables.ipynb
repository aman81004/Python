Variables : They are fundamental elements in programming used to store data that can be refrenced and manipulated in a program. In python , variables are created when you assign a value to them , and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
a = 100
```

```python
# Declaration of variable 
age = 21
name = "Aman"
height = 5.7
is_student = True

# Printing the variables
print("Age:", age)
print("Name:", name)
print("Height:", height)
print("Is Student:", is_student)
```

```python
# Naming Conventions
# Variable names should start with a letter or underscore
# Variable names can contain letters, digits, and underscores
# Variable names are case-sensitive

first_name = "Shaikh Mohammed"
lastName = "Fazlur Rahman"
_fullName = first_name + " " + lastName
print("Name:", _fullName)
```

```python
# Understanding Variable Types
age = 21
name = "Aman"
height = 5.7
is_student = True

print("Type of age:", type(age))
print("Type of name:", type(name))
print("Type of height:", type(height))
print("Type of is_student:", type(is_student))
```

```python
# Type checking and type conversion
age = 21
print("Type of age:", type(age))  # Output: <class 'int'>

age_str = str(age)
print("Age as string:", age_str)  # Output: "21"
print("Type of age_str:", type(age_str))  # Output: <class 'str
```

```python
# Dynamic Typing
variable = 10
print(variable, type(variable))  # Output: <class 'int'>
variable = "Hello"
print(variable, type(variable))  # Output: <class 'str'>
```

```python
# Simple Calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
