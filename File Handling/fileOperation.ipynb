File Operation - Read and write files

File handling is a crucial part of any programming lang. Python provides built in functions and methods to read from and write to files , both text and binary . This lesson will cover the basics of file handling , including reading and writing text files and binary files.

```python
### Read the whole file and print it out
with open('file.txt', 'r') as file:
    content = file.read()
    print(content)
    

        
```

```python
### Read the file line by line and print each line
with open('file.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
## Writing a file (overwriting existing content)

with open('file.txt','w') as file :
    file.write("Hello world.\n")
    file.write("I am new here.\n")
    
```

```python
## Write to a file (appending to existing content)
with open('file.txt','a') as file :
    file.write("This is an additional line.\n")
```

```python
## Writing a list of lines to a file
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
with open('file.txt', 'a') as file:
    file.writelines(lines)
```

```python
## Binary file writing
data = b'\x00\x01\x02\x03\x04'
with open('file.bin', 'wb') as file:
    file.write(data)
```

```python
## Reading in a binary file
with open('file.bin', 'rb') as file:
    binary_content = file.read()
    print(binary_content)
```

```python
## read the content from a file and write in another file 
with open('file.txt', 'r') as source_file:
    content = source_file.read()
    
with open('copy_file.txt', 'w') as dest_file:
    dest_file.write(content)
    
```
