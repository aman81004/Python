```python
## Create new directory
import os
new_directory = 'new_folder'
if not os.path.exists(new_directory):
    os.makedirs(new_directory)
    print(f"Directory '{new_directory}' created.")
```

```python
##List all files and directories in the current directory
items = os.listdir('.')
print("Files and directories in the current directory:",items)
```

```python
## Joining paths
file_path = os.path.join(new_directory, 'file.txt')
print("Full file path:", file_path)
```

```python
full_path = os.path.join(os.getcwd(),new_directory, 'file.txt')
print("Full file path:", full_path)
```

```python
## Checking if a path is a file or directory
path = 'file.txt'
if os.path.isfile(path):
    print(f"{path} is a file.")
elif os.path.isdir(path):
    print(f"{path} is a directory.")
else:
    print(f"{path} does not exist.")
    
```

```python
## Getting the file size
relative_path = 'file.txt'
absolute_path = os.path.abspath(relative_path)
if os.path.isfile(absolute_path):
    file_size = os.path.getsize(absolute_path)
    print(f"Size of '{relative_path}' is {file_size} bytes.")
else:
    print(f"'{relative_path}' does not exist.")
    
```

```python
## Getting the absolute path of a file
relative_path = 'file.txt'
absolute_path = os.path.abspath(relative_path)
print(f"Absolute path of '{relative_path}': {absolute_path}")
```
