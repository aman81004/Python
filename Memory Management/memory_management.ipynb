Memory Management 
Memory management in Python involves a combination of automatic garbage collection , reference counting and various internal optimizations to efficiency manage memory allocation and deallocation . Understanding these mechanisms can help developers write more efficient and robust applications.

Reference counting 

Reference counting is the primary method Python uses to manage memory. Each object in Python maintains a
count of references pointing to it. When the reference count drops to zero, the memory occupied by the object
is deallocated.

```python
import sys

a = []
## 2 ( one for a and one for the argument to getrefcount)
print(sys.getrefcount(a))
```

```python
b = a
## 3 ( one for a, one for b and one for the argument to getrefcount)
print(sys.getrefcount(a))
```

```python
del b 
## 2 ( one for a and one for the argument to getrefcount)
print(sys.getrefcount(a))
```

Garbage collection is the process of automatically freeing memory that is no longer in use by the program. In Python, garbage collection is handled by a built-in module called `gc`. The `gc` module provides functions for controlling the garbage collection process, such as enabling or disabling it, and for collecting and freeing memory.


```python
import gc
gc.collect()
```

```python
gc.disable()
```

```python
gc.collect()
```

```python
## Get garbage collection stats
print(gc.get_stats())
```

```python
## unreachable objects
print(gc.garbage)
```

Memory Management Best Practices

1.Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global
variables.
2Avoid Circular References: Circular references can lead to memory leaks if not properly managed.
Use Generators: Generators produce items one at a time and only keep one item in memory at a time,
making them memory efficient.
Explicitly Delete Objects: Use the del statement to delete variables and objects explicitly.
Profile Memory Usage: Use memory profiling tools like tracemalloc and memory_profiler to identify
memory leaks and optimize memory usage.


```python
import gc
from typing import Optional

class MyObject :
    def __init__(self,name):
        self.name = name 
        self.ref: Optional['MyObject'] = None
        print(f"{self.name} created")
        
    def __del__(self):
        print(f"{self.name} deleted")
        
        
## Create circular reference
obj1 = MyObject("Object 1")
obj2 = MyObject("Object 2")

obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2

## Manually trigger the garbage collection
gc.collect()
```

```python
## Generators for memory efficiency
## Generators allow you to iterate over large datasets without loading everything into memory at once.

def generate_numbers(n):
    for i in range(n):
        yield i
        
## Using the generator

for num in generate_numbers(100000):
    print(num)
    if num > 10:
        break
```

```python
## Profiling memory usage with tracemalloc

import tracemalloc

def create_list():
    return [i for i in range(100000)]

def main():
    tracemalloc.start()
    
    create_list()
    
    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')
    
    print("[ Top 10 ]")
    for stat in top_stats[:10]:
        print(stat)
        
```

```python
main()
```
