Function :
A function is a block of code that performs a specific task. Function help in organising code , reusing code and improving readability.

```python
## syntax
def function_name(parameters):
    # function body
    return parameters
```

```python
## odd even function
def odd_even(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
## calling the function
print(odd_even(24))
```

```python
## Funtion with multiple parameters
def add(a,b):
    return a + b
```

```python
result = add(5, 10)
print(result)
```

```python
## Default parameter
def greet(name):
    print(f"Hello, {name} Welcome to the paradise!")
    
greet("Alice")  # Output: Hello, Alice!
```

```python
## Variable length arguments
## positional and keyword arguments

def print_numbers(*aman):
    for numbers in aman:
        print(numbers)
```

```python
print_numbers(1, 2, 3, 4, 5)
```

```python
print_numbers(1, 2, 3, 4, 5,"Aman")
```

```python
## Keywords arguments
def print_details(**kwargs):
    for key , value in kwargs.items() :
        print(f"{key} : {value}")
```

```python
print_details(name="Aman", age=25, city="New York")
```
