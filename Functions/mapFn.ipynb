The map() Function

It applies a given function to all items in an input list and returns a map object . This is particularly useful for transforming data in a list comprehensively .

```python
numbers = [1,2,3,4,5,6]

squared = list(map(lambda x : x**2 , numbers))
print(squared)
```

```python
## map multiple iterables
numbers1 = [1,2,3,4,5,6]
numbers2 = [7,8,9,10,11,12]

added_numbers = map(lambda x,y : x + y , numbers1,numbers2)
print(list(added_numbers))
```

```python
## map to convert list of strings to integers

str_numbers = ['1','2','3','4','5']
int_numbers = list(map(lambda x : int(x) , str_numbers))
print(int_numbers)
```

```python
words = ['apple','banana','cherry']
upper_words = list(map(str.upper , words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name' : 'Alice' , 'age' : 30},    
    {'name' : 'Bob' , 'age' : 25},    
    {'name' : 'Charlie' , 'age' : 35}
]

names = list(map(get_name , people))
print(names)
```
