Understanding Exceptions
exception handling in python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This lesson will cover the basics of exception , including how to use try ,except ,else and finally blocks .


What are Exceptions?
Exceptions are events that disrupt the normal flow of program. They occur when an error is encountered during program exception . Common exceptions include :
    - zero division error : dividing by zero
    - file not found error : file not found
    - value error : invalid values
    - type error : invalid type 

```python
## Exception try 
try:
    a = b
except NameError as exception:
    print("The variable has not been assigned",exception)
    
```

```python
try:
    result = 1 / 0
except ZeroDivisionError as exception:
    print("You cannot divide by zero", exception)
    print("Please provide a non-zero denominator")
```

```python
try :
    num  = int(input("Enter a number: "))
    result = 10 / num
except ValueError as exception:
    print("Please enter a valid integer", exception)

except ZeroDivisionError as exception:
    print("You cannot divide by zero", exception)
    print("Please provide a non-zero denominator")
    
except Exception as exception:
    print("An error occurred", exception)
```

```python
## try , except , else and finally
try:
    num  = int(input("Enter a number: "))
    result = 10 / num
except ValueError as exception:
    print("Please enter a valid integer", exception)
except ZeroDivisionError as exception:
    print("You cannot divide by zero", exception)
    print("Please provide a non-zero denominator")
except Exception as exception:
    print("An error occurred", exception)
    
else:
    print("The result is:", result)
finally:
    print("This block will always execute")
```
