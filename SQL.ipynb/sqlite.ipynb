SQL and SQLite
Sql is a standard lang for managing and manipulating relational databases. Sqlite is a self contained , serverless and zero configuration databases engine that is widely used for embeddded databases systems . In this lesson , we will cover the basics of SQL and SQLite , includng creating databases , tables and performing various SQL operations .

```python
import sqlite3
```

```python
## Connect to the SQLite database (or create it if it doesn't exist)
connection = sqlite3.connect('example.db')
connection 
```

```python
cursor = connection.cursor()
```

```python
## Create a table
cursor.execute('''
               Create Table IF NOT EXISTS employees(
                id INTEGER PRIMARY KEY,
                name TEXT NOT NULL,
                age INTEGER,
                department TEXT
               )
    ''')

## commit the changes to the database
connection.commit()
```

```python
cursor.execute('''
               Select * from employees
    ''')
```

```python
## Insert some data into the table
cursor.execute('''
                Insert INTO employees (name, age, department) VALUES
                ('Alice', 30, 'HR'),
                ('Bob', 25, 'Engineering'),
                ('Charlie', 35, 'Sales')
     ''')

## commit the changes to the database
connection.commit()
```

```python
## Query the data
cursor.execute('''
               Select * from employees
    ''')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Update data
cursor.execute('''
                Update employees
                Set age = 31
                Where name = 'Alice'
    ''')
connection.commit()
```

```python
cursor.execute('''
               Select * from employees
    ''')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Delete data
cursor.execute('''
                Delete from employees
                Where name = 'Bob'
    ''')
connection.commit()
```

```python
cursor.execute('''
               Select * from employees
    ''')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## Working with sales data
## connect to the database
connection = sqlite3.connect('sales.db')
cursor = connection.cursor()

## Create a table for sales data
cursor.execute('''
               Create Table IF NOT EXISTS sales(
                id INTEGER PRIMARY KEY,
                product TEXT NOT NULL,
                sales INTEGER,
                region TEXT
               )
    ''')

## Insert some sales data
cursor.execute('''
                Insert INTO sales (product, sales, region) VALUES
                ('Product A', 100, 'North'),
                ('Product B', 150, 'South'),
                ('Product C', 200, 'East')
     ''')

## commit the changes to the database
connection.commit()

## Query the sales data
cursor.execute('''
               Select * from sales
    ''')

rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
## Close the database connection
connection.close()
```
