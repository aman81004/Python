Syntax : It refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a lang. In simpler terms , syntax is about the correct arrangement of words and symbols in a code.

Semantics : It refers to the meaning or the interpretation of the symbols, characters and commands in a lang. It is about what the code is supposed to do when it runs.

```python
# Basic syntax Rules in Python
# Case Sensitivity : Python is case-sensitive
name = "Aman"
print(name)
```

Indentation : It is used to define the structure and heirarchy of the code. Unlike many other programming lang that use braces {} to delimit blocks of code,Python uses indentation to determine the grouping of statements within a block must be idented at the same level.

```python
# Indentation : Python uses indentation to define blocks of code
age = 21
if age>18:
    print("You are an adult.")
else:  
    print("You are a minor.")
```

```python
# Line continuation : You can use a backslash (\) to continue a statement on the next line
total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)
```

```python
# Multipe statements in a single line : You can use a semicolon (;) to separate multiple statements on a single line
a = 5; b = 10; print(a + b)
```

```python
# Understanding Semantics in Python
# Variable Naming : Variable names must start with a letter or underscore and can contain letters, digits, and underscores
age = 21
name = "Aman"
```

```python
type(age)
type(name)
```

```python
#Type inference : Python automatically infers the data type of a variable based on the assigned value
variable = 10
print(type(variable))  # Output: <class 'int'>
variable = "Hello"
print(type(variable))  # Output: <class 'str'>
```
